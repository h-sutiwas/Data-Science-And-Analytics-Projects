# **Import libraries and dependencies**

```python
# Import libraries
import pandas as pd
from sqlalchemy import create_engine
# import pyodbc

import os
import glob
from dotenv import load_dotenv

import warnings
warnings.filterwarnings("ignore")
```

```python
load_dotenv()

db_user = os.environ.get('DB_USER')
db_password = os.environ.get('DB_PASSWORD')
db_server = os.environ.get('DB_SERVER')
db_name = os.environ.get('DB_NAME')

connection_string = f"mssql://{db_server}/{db_name}?driver=ODBC+DRIVER+17+FOR+SQL+SERVER"
engine = create_engine(connection_string, echo=True)
conn = engine.connect()

# conn_str = f"DRIVER={{ODBC Driver 17 for SQL Server}};SERVER={db_server};DATABASE={db_name};UID={db_user};PWD={db_password}"
# conn = pyodbc.connect(conn_str)

# conn = pyodbc.connect(f"DSN=sqlREMOTEserver;UID={db_user};PWD={db_password}")

print(f"Server: {db_server}, DB: {db_name}")
```

```python
csv_files = glob.glob(os.getcwd() + "/*.csv")

for file in csv_files:
    df = pd.read_csv(file)
    df.to_sql(f'{file[104:-4]}', con=conn, if_exists='replace', index=False)
```
